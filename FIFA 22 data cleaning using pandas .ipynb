```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_csv = r"/Users/miguelbrown/Desktop/Python pandas/FIFA_players_22.csv"
df = pd.read_csv(file_csv)
df.head(5)
```

```python
df= df.drop_duplicates()
df.head(5)
```

```python
# display all colums in df
print(df.columns.tolist())
```

```python
df= df.drop(columns=['player_url','club_loaned_from','weak_foot','real_face','player_tags','player_face_url','club_logo_url','club_flag_url','nation_logo_url','nation_flag_url'])
df
```

```python
df = df.replace('NaN','')
df = df.fillna('')
df
```

```python
df.set_index(df.columns[0])
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(df.corr(), annot = True)

plt.show()
```

```python
df.select_dtypes(include='number')
```

```python
df['wage_eur'] = pd.to_numeric(df['wage_eur'])
```

```python
plt.figure(figsize=(7, 5))
ax = sns.scatterplot(x =df['age'], y = df['wage_eur'])
plt.xlabel("age") 
plt.ylabel("Wage")
plt.title("age & wage", fontsize = 18)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,5))
df.groupby("club_position")["wage_eur"].max().plot.bar()
plt.xlabel("club position") 
plt.ylabel("Wages in EUR")
plt.title("Wages by club position", fontsize = 15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
df.groupby("club_joined")["wage_eur"].max().plot.line()
plt.xlabel("club position") 
plt.ylabel("Wages in EUR")
plt.title("Wages by club position", fontsize = 15)
plt.show()
```
